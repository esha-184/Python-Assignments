## Assignent Question 1
Using nested for loop to print multiplication table of 5.

```python
for table in [5]:
  print("Table of" , table)
  for i in range(1,11):
    print(table,"x",i,"=", table*i)
    print()
```

## Assignent Question 2
Store a password in variables. Use a for loop to count how many digits and letters are present in the password.

```python
password=input("Enter a password:")
digits=""
letters=""

for character in password:
  if character .isdigit():
    digits += character
  elif character .isalpha():
    letters += character
print("Letters are :", letters)
print("Total letters:", len(letters))
print("Numbers are :", digits)
print("Total numbers:", len(digits))
```

##Assignent Question 3
Keep asking the user to enter item prices. Stop when the user enters 0, then display the total bill.

```python
total = 0
price = int(input("Enter price of items(0 to stop):"))
while price != 0:
  total += price
  price = float(input("Enter price of items(0 to stop):"))
print("Total bill is:", total)
```

## Assignment Question 4
There are 3 users,each user gets 2 attempts to enter the correct password(python456).

```python

correct_password = "Python123"

attempt = 1

while attempt <= 2:
    password = input("Enter password: ")

    if password == correct_password:
        print("Login Successful")
        break

    else:
        print("Wrong Password")
        attempt += 1

if attempt > 2:
    print("Failed to Login")
```
